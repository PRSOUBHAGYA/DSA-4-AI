## 1. Builtin Datatypes

### 1.1. Numerical Types

```python
num = 8
type(num)
```

```python
num = 2.
type(num)
```

```python
num = 0 + 0j
type(num)
```

### 1.2. Sequence Types

```python
value = 'Hello world'
value_2 = "Hello World"
value_3 = """ hello
world"""
type(value), type(value_2), type(value_3)
```

```python
val_list = ['Hello','World',1,2+6j]
print(type(val_list))
```

```python
val_tuple = (1,5,9)
val_tuple_2 = 1,0,9
type(val_tuple),type(val_tuple_2)
```

```python
tuple_1 = 4,
tuple_2 = (4)
type(tuple_1), type(tuple_2)
```

```python
val = range(0,5)
type(val)
```

```python
list(val)
```

### 1.6. None Type

## 2. Control Structure

### 2.1. if-else

```python
a = 7
if a > 8:
    print("small than 8")
else:
    print("greater or equal to 8")
```

### 2.2. if-elif-else

```python
a = 8
if a > 8:
    print("smaller than 8")
elif a < 8 :
    print("greater than 8")
else:
    print("Both are equal")
```
