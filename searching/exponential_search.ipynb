```python
def binary_search(arr,key):
    low = 0
    high = len(arr)-1

    while low<= high:
        mid = low + (high-low)//2
        if arr[mid] == key:
            return True
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return False
```

```python
def exponential_search(arr,key):
    if arr[0] == key:
        return True
    i = 1
    while i<= len(arr) and arr[i]< key:
        i *= 2

    low = i//2
    high = min(i,len(arr)-1)
    print(low,high)
    return binary_search(arr[low:high+1],key)
```
