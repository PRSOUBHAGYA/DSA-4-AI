```python
def binary_search(arr, key):
    low = 0
    high = len(arr)-1

    while low <= high:
        mid = low + (high-low)//2
        if arr[mid] < key:
            low = mid + 1
        elif arr[mid] > key:
            high = mid - 1
        else:
            return True

    return False
```

```python
binary_search([9,10,56,89],1)
```
