# List

## 1. 1D list Operation 

```python
# Declaration
list1 = [1,0,8,'Hello',3.8,2+1j]
list1
```

```python
# append
list1.append(True)
list1
```

```python
# insert
list1.insert(2,['new ele'])
list1
```

```python
# extend
list1.extend(['H',-90,1.0,False])
list1
```

```python
# add operator to extend
list1+['I','WIll','win']
```

```python
# remove
list1.remove(-90)
list1
```

```python
# pop
list1.pop()
```

```python
list1.pop(2)
```

```python
list1
```

```python
# clear
list1.clear()
list1
```

```python
# indexing
list2 = ['love',True,2.00000009,False,'kill',False] 
list2.index(False)
```

```python
# count
list2.count(False)
```

```python
#sort
list2.sort()
```

```python
list3 = [9,89,2,90,123]
list3.sort()
list3
```

```python
# reverse
list2.reverse()
list2
```
