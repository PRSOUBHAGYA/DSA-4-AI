# Character Count

```python
def get_char_count(text):
    char_dict = dict()
    for each in text:
        char_dict[each] = char_dict.get(each,0)+1

    return char_dict
```

```python
get_char_count('Hello')
```

# First Non repeating character

```python
def get_first_non_repeating_char(text):
    char_count_dict = dict()
    for each in text:
        char_count_dict[each] = char_count_dict.get(each,0)+1


    for key,value in char_count_dict.items():
        if value == 1:
            return key
    
    return None
```

```python
get_first_non_repeating_char('hhis is Soubhagya')
```

# Two sum Problem

```python
def get_sum_pair(seq,target):
    seen = set()
    for each in seq:
        if target-each in seen:
            return (target-each, each)
        seen.add(each)
    return None
```

```python
get_sum_pair([1,8,9,4,5,6],14)
```
