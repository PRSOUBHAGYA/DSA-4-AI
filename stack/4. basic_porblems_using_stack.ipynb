# Reverse String Using Stack

```python
from collections import deque
def reverse_string(val_str):
    stack = deque()

    for each in val_str:
        stack.append(each)

    result = ""
    while stack:
        result += stack.pop()

    return result
```

```python
reverse_string('Hello')
```

```python
reverse_string('KataK')
```

# Check Balanced Paranthesis

```python
from collections import deque
def check_balanced_parantheisis(expression):
    stack = deque()
    pairs = {
        ')':'(',
        '}':'{',
        ']':'['
    }

    for each in expression:
        if each in '[{(':
            stack.append(each)
        if each in ']})':
            if not stack or stack.pop() != pairs[each]:
                return False

    return len(stack) == 0
```

```python
check_balanced_parantheisis('(9+0+(8-9*[1-0]+{(12-8)/9})))')
```
