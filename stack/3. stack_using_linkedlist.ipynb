# Stack Using Linkedlist

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top == None

    def display(self):
        if self.is_empty():
            return "Nothing to display as stack is empty"
        current = self.top
        while current != None:
            print(current.data,end='\n')
            current = current.next
        
    def push(self,data):
        node1 = Node(data)
        node1.next = self.top
        self.top = node1

    def pop(self):
        if self.is_empty():
            return "Nothing to pop as stack is empty"

        value = self.top.data
        self.top = self.top.next

    def peek(self):
        if self.is_empty():
            return "Nothing to peek as stack is empty"

        return self.top.data

        
```

```python
stack = Stack()
stack.push(7)
stack.push(6)
stack.push(5)
stack.push(4)
stack.push(3)
stack.display()
```

```python
stack.pop()
stack.display()
```
