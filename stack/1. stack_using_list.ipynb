# Implementing stack using list

```python
class Stack:
    def __init__(self):
        self.stack = []
        self.length = 0

    def is_empty(self):
        return len(self.stack) == 0

    def display(self):
        for each in self.stack[::-1]:
            print(each,end='\n')

    def push(self,data):
        self.stack.append(data)
        self.length += 1

    def pop(self):
        if self.is_empty():
            return None
        item = self.stack.pop()
        self.length -= 1
        return item
    
    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]

            
```

```python
st = Stack()
st.push(4)
st.push(3)
st.push(1)
st.pop()
st.display()
st.peek()
```
