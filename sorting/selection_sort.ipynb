```python
def selection_sort(arr):
    if len(arr)<= 1:
        return arr
    for i in range(len(arr)-1):
        min_inx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_inx]:
                min_inx = j
        arr[i],arr[min_inx] = arr[min_inx],arr[i]

    return arr
```

```python
selection_sort([2,0,8,4])
```
