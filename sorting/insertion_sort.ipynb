```python
def insertion_sort(arr):
    if len(arr)<= 1:
        return arr

    for i in range(1,len(arr)):
        key = arr[i]
        j = i-1
        while j>=0 and arr[j]>key:
            arr[j+1] = arr[j]
            j = j-1

        arr[j+1] = key
    return arr
```

```python
insertion_sort([1,0,9,6])
```
