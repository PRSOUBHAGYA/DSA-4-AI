```python
def bubble_sort(arr):
    if len(arr) == 0:
        print("Array is empty! Nothing to sort!!")
        return None
    if len(arr) == 1:
        return arr
    
    for j in range(len(arr)):
        swapped = False
        for i in range(len(arr)-j-1):
            if arr[i] > arr[i+1]:
                arr[i],arr[i+1] = arr[i+1],arr[i]
                swapped = True
        if not swapped:
            break
    return arr



    
```

```python
buuble_sort([7,9,0,5,2])
```
