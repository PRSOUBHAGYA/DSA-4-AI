```python
class Node:
    def __init__(self,data):
        self.data = data
        self.prev = None
        self.next = None
```

```python
class DoubleLinkedList:
    def __init__(self):
        self.head = None
    
    def get_length(self):
        count = 0
        current = self.head
        while current != None:
            count +=1
            current = current.next
        return count

    def display(self):
        current = self.head
        while current != None:
            print(current.data,end=' ')
            current = current.next
    
    def add_node_at_beginning(self,data):
        node1 = Node(data)
        if self.head == None:
            self.head = node1
        else:
            node1.next = self.head
            self.head.prev = node1
            self.head = node1
            

    def add_node_at_end(self,data):    
        node1 = Node(data)
        if self.head == None:
            self.head = node1
        else:
            current = self.head
            while current.next!= None:
                current = current.next
            current.next = node1
            node1.prev = current

    def add_node_in_middle(self,data,pos):
        if pos <=0  or pos > self.get_length()+1:
            print('Invalid position')
            return -1
        elif pos == 1:
            self.add_node_at_beginning(data)
        elif pos == self.get_length()+1:
            self.add_node_at_end(data)
        else:
            current = self.head
            for _ in range(1,pos-1):
                current = current.next
            node1 = Node(data)
            node1.prev = current
            node1.next = current.next
            current.next.prev = node1
            current.next = node1




    
```

```python
dl = DoubleLinkedList()
dl.add_node_at_beginning(12)
dl.add_node_at_beginning(22)
dl.add_node_at_end(32)
dl.display()
```
