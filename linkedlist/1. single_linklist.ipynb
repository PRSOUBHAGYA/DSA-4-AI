# Single LinkedList

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def get_length(self):
        count = 0
        current = self.head

        while current!= None:
            count += 1
            current = current.next
        return count

    def display(self):
        if self.head is None:
            print("Linedlist is empty!!")
            return
        current = self.head
        while current != None:
            print(current.data,end=' ')
            current = current.next

    def add_node_in_beginning(self,data):
        current = self.head
        node = Node(data)
        self.head = node
        if self.head != None:
            node.next = current

    def add_node_in_end(self,data):
        current = self.head
        if current is None:
            self.add_node_in_beginning(data)
        else:
            node = Node(data)
            while current.next != None:
                current = current.next

            current.next = node

    def add_node_in_middle(self,data,pos):
        if pos<0 or pos > self.get_length()+1:
            print("Invalid position")
        elif pos == 1:
            self.add_node_in_beginning(data)
        else:
            current = self.head
            for i in range(1,pos-1):
                current=current.next
            node1 = Node(data)
            node1.next = current.next
            current.next = node1

    def search_value(self,key):
        if self.head==None:
            print('No elements are there')
            return -1
        pos = 0
        current = self.head
        while current != None:
            if current.data == key:
                return pos
            current = current.next
            pos += 1
        return -1
    
    def deletion_at_beginning(self):
        if self.head==None:
            print("Delete is not possible due to no elements!!")
            return -1
        elif self.head.next==None:
            self.head = None
        else:
            self.head = self.head.next

    def deletion_at_end(self):
        if self.head==None:
            print("Delete is not possible due to no elements!!")
            return -1
        elif self.head.next==None:
            self.head = None
        else:
            current = self.head
            while current.next.next != None:
                current = current.next
            current.next = None

    def deletion_at_middle(self,pos):
        if pos < 0 or pos > self.get_length():
            print('deletion is not possible')
            return -1
        elif pos == 1:
            self.head=self.head.next
        else:
            current = self.head
            for _ in range(1,pos-1):
                current = current.next
            current.next = current.next.next

    def find_middle_element(self):
        pos = self.get_length()//2
        current = self.head
        for _ in range(pos):
            current = current.next
        return current.data

    def find_sum(self):
        sum = 0
        if self.head == None:
            return sum
        current = self.head
        while current != None:
            sum += current.data
            current = current.next
        return sum

    def find_max_min_element(self):
        _min = 99999999999
        _max = -99999999999

        if not self.head:
            return -1
        
        current = self.head
        while current:
            if current.data < _min:
                _min = current.data
            if current.data > _max:
                _max = current.data
            
            current = current.next

        return {'max':_max,'_min':_min}

            
```
