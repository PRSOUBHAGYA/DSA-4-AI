# Single LinkedList

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def get_length(self):
        count = 0
        current = self.head

        while current!= None:
            count += 1
            current = current.next
        return count

    def display(self):
        if self.head is None:
            print("Linedlist is empty!!")
            return
        current = self.head
        while current != None:
            print(current.data,end=' ')
            current = current.next

    def add_node_in_beginning(self,data):
        current = self.head
        node = Node(data)
        self.head = node
        if self.head != None:
            node.next = current

    def add_node_in_end(self,data):
        current = self.head
        if current is None:
            self.add_node_in_beginning(data)
        else:
            node = Node(data)
            while current.next != None:
                current = current.next

            current.next = node

    def add_node_in_middle(self,data,pos):
        if self.get_length() >= pos:
            if pos == 1:
                node = Node(data)
                current = self.head
                self.head = node
                node.next = current
            elif pos == self.get_length():
                current = self.head
                while current.next.next!=None:
                    current = current.next
                node = Node(data)
                node.next = current.next
                current.next=node
            else:
                count = 1
                current = self.head
                while count <= pos-2:
                    current = current.next
                    count += 1
                node = Node(data)
                node.next = current.next
                current.next = node
        else:
            print("LinkedList is not having that position to insert.")
```

```python
node1 = Node(5)
linkedlist_1 = LinkedList()
print(linkedlist_1.get_length())
linkedlist_1.add_node_in_beginning(6)
linkedlist_1.add_node_in_beginning(5)
linkedlist_1.add_node_in_beginning(0)
linkedlist_1.add_node_in_end(10)
linkedlist_1.get_length()
```
