# Prime Number

```python
def is_prime(val):
    flag = True
    for i in range(2,val):
        if val % i == 0:
            flag = False
            break
    if flag:
        return 'Prime'
    else:
        return 'Not a Prime'
```

```python
is_prime(19)
```

# Fibbonanci Series

```python
def find_fibonnaci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return find_fibonnaci(n-1) + find_fibonnaci(n-2)

for i in range(8):
    print(find_fibonnaci(i),end=' ')
        
```

```python
def fibonacci(n):
    a,b = 0,1
    print(0, end=' ')
    for each in range(n-1):
        print(b,end=' ')
        a,b = b,a+b

fibonacci(8)
```

# ArmStrong Number

```python
import math
def is_armstrong(n):
    org = n
    len_n = len(str(n))
    sum_n = 0
    while n != 0:
        sum_n += math.pow(n%10,len_n)
        n = int(n/10)

    if sum_n == org:
        print('Yes')
    else:
        print('No')
    
```

```python
is_armstrong(1634)
```

# Factorial

```python
def find_factorial(n):
    res = 1
    for each in range(1,n+1):
        res *= each
    return res
```

```python
find_factorial(5)
```

```python
def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)
```

```python
fact(5)
```

# Anagram

```python
def is_anagram(str1,str2):
    if sorted(str1)==sorted(str2):
        return True
    else:
        return False
```

```python
is_anagram('listen','silent')
```

```python
from collections import Counter
def is_anagram_2(str1,str2):
    if Counter(str1) == Counter(str2):
        return True
    else:
        return False
```

```python
is_anagram_2('silent','listen')
```
