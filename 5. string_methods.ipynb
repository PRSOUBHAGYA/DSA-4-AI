# Basic String Methods

```python
# lower
s = 'SoUbHagYa pRadHan'
s.lower()
```

```python
# len
len(s)
```

```python
# count
s.count('a')
```

```python
s.count('m')
```

```python
# upper
s.upper()
```

```python
# title
s.title()
```

```python
# capitalize
s.capitalize()
```

```python
# swapcase
s.swapcase()
```

```python
# strip
s = ' I love my country India     '
s.strip()
```

```python
# lstrip
s.lstrip()
```

```python
# rstrip
s.rstrip()
```

```python
s = '## hello ##&'
s.strip('#').strip('&').strip('#').strip()
```

```python
# replace
s = 'Soubhagya'
s.replace('a','i')
```

```python
s.replace('a','i',1)
```

```python
#split
s = 'Hello Mr. Sharma! How are you?'
s.split(' ')
```

```python
s.split(' ',maxsplit=1)
```

```python
s = 'Hello! Mr. Sharma! How are you?'
s.split('!')
```

```python
# join
name_list = ['Ram', 'Shyam', 'Kam', 'Naam']
' -> '.join(name_list)
```
