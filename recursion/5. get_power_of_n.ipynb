```python
def get_power(n,k):
    if k==0:
        return 1
    return n*get_power(n,k-1)
```

```python
get_power(-2,4)
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def get_fast_power(n,k):
    if k==0:
        return 1

    if k==1:
        return n

    if k%2==0:
        half = get_fast_power(n,k//2)
        return half*half

    return n*get_fast_power(n,k-1)
```

```python
get_fast_power(3,90)
```
