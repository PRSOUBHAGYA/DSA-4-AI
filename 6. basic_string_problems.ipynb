# Basic String Problems

## 1. Count numbers of Vowels, Constants

```python
s = '_hell o123$world'
type(s)
```

```python
def get_count_vowels_constants(s):
    vowels = 'aeiouAEIOU'
    count_vow = 0
    count_con = 0

    for each in s:
        if each in vowels:
            count_vow += 1
        elif each.isalpha():
            count_con += 1

    return count_vow, count_con
```

```python
vow, con = get_count_vowels_constants(s)
f'There are {vow} number of vowels and {con} number of consonants.'
```

## 2. Count Frequency of Characters

```python
def get_frequency_of_chars(s):
    count_dict = {}

    for each in s:
        if each in count_dict:
            count_dict[each] += 1
        else:
            count_dict[each] = 1

    return count_dict
```

```python
get_frequency_of_chars('Banana')
```

```python
def get_frequency_of_chars_using_get(s):
    count_dict = {}
    
    for each in s:
        count_dict[each] = count_dict.get(each,0)+1
    return count_dict
```

```python
get_frequency_of_chars_using_get('Banana')
```

## 3. pallindrome

```python
def is_pallindrome(s):
    return s==s[::-1]
```

```python
is_pallindrome('katak')
```

## 4. anagram

```python
def check_anagram(base, target):
    return sorted(base) == sorted(target)
```

```python
check_anagram('listen','silent')
```

```python
def check_anagram_using_count_frequency(base,target):
    return Counter(base) == Counter(target)
```

```python
check_anagram_using_count_frequency('liisten','silenti')
```

```python
def check_anagram_using_single_count_dict(base, target):
    if len(base) != len(target):
        return False

    count_dict={}
    for char1, char2 in zip(base,target):
        count_dict[char1] = count_dict.get(char1,0) + 1
        count_dict[char2] = count_dict.get(char2,0) - 1
    
    return all([each == 0 for each in count_dict.values()])
```

```python
check_anagram_using_single_count_dict('liisten','silenti')
```
