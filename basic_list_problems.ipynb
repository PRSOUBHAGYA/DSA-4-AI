# Basic List Problems

## 1. Find the largest number in the list without max()

max([1,9,23,9]) -> 23

```python
def find_max(items_list):
    max_num = items_list[0]

    for each in items_list[1:]:
        if each >= max_num:
            max_num = each

    return max_num
```

```python
find_max([1,90,8,7,66])
```

## 2. Count occurance of element in a list without count()

count([1,0,0,9,6,0],0) -> 3

```python
def count_occurance_ele(items_list,value):
    count = 0

    for each in items_list:
        if each == value:
            count += 1

    return count
```

```python
count_occurance_ele(['A','Z','C','F','Z','B'], 'Z')
```

## 3. Remove Duplicates from List without set()

list(set(1,8,9,0,0,9,4,5,1)) - [0,1,4,5,8,9]

```python
def remove_duplicates(items_list):
    unique_list = []
    
    for each in items_list:
        if each not in unique_list:
            unique_list.append(each)

    return unique_list
```

```python
remove_duplicates(['A','Z','C','F','Z','B'])
```

## 4. Reverse a list without reverse() or [::-1]

[1,9,0,7,8][::-1] -> [8,7,0,9,1]

### 4.1. Using For loop

```python
def reverse_list(items_list):
    result_list = []
    n = len(items_list)
    for each in range(n-1,-1,-1):
        result_list.append(items_list[each])

    return result_list
```

```python
reverse_list(['A', 'Z', 'C', 'F', 'B'])
```

### 4.2. Using while loop

```python
def reverse_list(items_list):
    result_list = []
    n = len(items_list)
    i = n-1
    while i >= 0:
        result_list.append(items_list[i])
        i -= 1

    return result_list
```

```python
reverse_list(['A', 'Z', 'C', 'F', 'B'])
```
