```python
from collections import deque
class Queue:
    def __init__(self):
        self.queue = deque()

    def is_empty(self):
        return len(self.queue) == 0

    def display(self):
        if not self.is_empty():
            print(self.queue)
        else:
            print('Queue is empty')
            return None

    def enqueue(self,data):
        self.queue.append(data)
        print(f'Inserted {data} to the queue!')

    def dequeue(self):
        if self.is_empty():
            print("Nothing to pop as queue is empty!")
            return None
        item = self.queue.popleft()
        return item

    def front(self):
        if self.is_empty():
            print('Queue is empty!!')
            return None
        
        return self.queue[0]

    def rear(self):
        if self.is_empty():
            print('Queue is empty!!')
            return None

        return self.queue[-1]
```

```python
q = Queue()
q.display()
```

```python
q.enqueue(3)
q.enqueue(5)
q.enqueue(6)
#q.display()
```
