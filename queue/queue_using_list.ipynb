```python
class queue:
    def __init__(self):
        self.queue = []
    
    def is_empty(self):
        return len(self.queue) == 0

    def display(self):
        if not self.is_empty():
            return self.queue
        else:
            print("Noting to show; queue is empty")
            return

    def enqueue(self,data):
        self.queue.append(data)

    def dequeue(self):
        if self.is_empty():
            print('queue is empty; nothing to deque!')
            return
        ele = self.queue.pop(0)
        return ele

    def peekfront(self):
        return self.queue[0]

    def peekrear(self):
        return self.queue[-1]
    
```

```python
q = queue()
q.enqueue(6)
q.enqueue(7)
q.enqueue(8)
q.dequeue()
q.display()
```
