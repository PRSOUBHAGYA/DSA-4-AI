```python
from collections import deque
```

```python
class Queue:
    def __init__(self):
        self.stack1 = deque()
        self.stack2 = deque()

    def is_empty(self):
        return len(self.stack1) == 0 and len(self.stack2) == 0

    def display(self):
        if self.is_empty():
            print("Queue is Empty!!")
            return None
        
        print(list(self.stack2)[::-1]+ list(self.stack1))

    def enqueue(self,data):
        self.stack1.append(data)

    def dequeue(self):
        if self.is_empty():
            print('Queue is Empty !')
            return None
        
        if len(self.stack2) == 0:
            while len(self.stack1) != 0:
                self.stack2.append(self.stack1.pop())
        
        item = self.stack2.pop()

        return item

    
```

```python
q = Queue()
q.display()
```

```python
q.enqueue(4)
q.enqueue(5)
q.enqueue(3)
q.dequeue()
q.enqueue(9)
q.display()
```
