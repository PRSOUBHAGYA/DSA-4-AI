```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.head = None
        self.tail = None

    def is_empty(self):
        return self.head == None

    def display(self):
        if self.is_empty():
            print("Nothing to display as queue is empty!")
            return None

        current = self.head
        while current != None:
            print(current.data,end=' ')
            current = current.next

    def enqueue(self,data):
        node = Node(data)
        if self.is_empty():
            self.head = node
            self.tail = node
        else:
            self.tail.next = node
            self.tail = node

    def dequeue(self):
        if self.is_empty():
            print('Nothing to pop as queue is empty')
            return None
        
        item = self.head.data
        self.head = self.head.next

        if self.head == None:
            self.tail = None

        return item

    def front(self):
        if not self.is_empty():
            return self.head.data
        else:
            print('Queue is empty!')
            return None

    def rear(self):
        if not self.is_empty():
            return self.tail.data
        else:
            print('Queue is empty!')
            return None
            
```

```python
q = Queue()
q.display()
```

```python
q.enqueue(3)
q.enqueue(5)
q.enqueue(6)
```

```python
q.display()
```

```python
q.dequeue()
```
