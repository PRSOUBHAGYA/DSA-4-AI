# Reverse first k elements

```python
from collections import deque
def reverse_first_k_in_queue(q,k):
    stack = deque()
    res = deque()
    for _ in range(k):
        stack.append(q.popleft())

    while stack:
        res.append(stack.pop())
    
    while q:
        res.append(q.popleft())

    return list(res)
```

```python
q = deque()
for i in range(1,6):
    q.append(i)

reverse_first_k_in_queue(q,2)
```

# Wierd ticket counter

```python
from collections import deque
def ticket_counter(n,k):
    if n == 1:
        return 1

    q = deque()
    for i in range(1,n+1):
        q.append(i)
    
    while True:
        for _ in range(k):
            q.popleft()
            if len(q) == 1:
                return q[0]
        for _ in range(k):
            q.pop()
            if len(q) == 1:
                return q[0]


        

```
