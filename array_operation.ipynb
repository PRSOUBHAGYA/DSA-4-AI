# Array

## 1. Array

```python
import array
```

```python
arr1 = array.array('i',[12,9,4,98,9])
type(arr1)
```

```python
# Insertion
arr1.insert(2,-2)
arr1
```

```python
# deletion
arr1.remove(98)
arr1
```

```python
# updation
arr1[2] = 2
arr1
```

```python
# sorting
array.array("i",sorted(arr1))
```

```python
# merging
arr2 = array.array("i",[11,12,13,90])
arr1+arr2
```

```python
# splitting
arr1[:2]
```

## 2. Numpy Array

```python
import numpy as np
```

```python
arr1 = np.array([12,8,5,0])
arr1
```

```python
# Insertion
np.insert(arr1,-2,23)
```

```python
# Deletion
np.delete(arr1,1)
```

```python
# Updation
arr1[0]=1
arr1
```

```python
# Sorting
np.sort(arr1)
```

```python
# merging
arr2 = np.array([99,98,76])
np.concatenate([arr1,arr2])
```

```python
# splitting
np.split(arr1,2)
```
